# Fetch URNs from ID in bibliography

```python
from bs4 import BeautifulSoup
import requests
```

```python
import re
```

```python
file_location = '/gdrive/My Drive/norbok_nasjonal.xml'
```

```python
with open(file_location) as f:
    text = f.read()
```

```python
text[:300]
```

```python
ids = [x.split('\n')[1].strip()[:-1] + '2' for x in re.findall('controlfield tag="001".[^<]+', text)]
```

```python
def get_urn(ident):
    nbapi = """https://api.nb.no/catalog/v1/items?q=oaiid:oai:nb.bibsys.no:{mmsid}""".format(mmsid=ident)
    u = requests.get(nbapi).json()
    try:
        res = u['_embedded']['items'][0]['metadata']['identifiers']['urn']
    except:
        res = 'none'
    return res
```

```python
urn_mapping = {ident:get_urn(ident) for ident in ids}
```

```python
with open('/gdrive/My Drive/urn_mapping.json','w') as f:
    json.dump(urn_mapping, f)
```
